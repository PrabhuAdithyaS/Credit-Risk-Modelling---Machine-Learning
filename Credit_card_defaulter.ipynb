This dataset is highly unbalanced
0 -> Normal Transaction
1 -> Fraudlent Transaction

```python
#Separating the data for analysis
legit_df = input_df[input_df['Class']==0]
fraud_df = input_df[input_df['Class']==1]
```

```python
print(legit_df.shape)
print(fraud_df.shape)
```

```python
#Amount column statistics - legit
legit_df['Amount'].describe()
```

```python
#Amount column statistics - fraud
fraud_df['Amount'].describe()
```

```python
legit_df = legit_df.dropna(subset=['Amount'])
```

```python
legit_df.shape
```

```python
input_df.groupby('Class').mean()
```

Build a sample dataset containing similar distribution of legit transactions and Fraudlent transactions - This is done in order to overcome the unbalanced dataset

```python
legit_sample = legit_df.sample(n=492)
```

```python
legit_sample.describe()
```

```python
legit_sample.shape
```

```python
new_sample_dataset = pd.concat([legit_sample,fraud_df],axis=0)
new_sample_dataset.shape
```

```python
new_sample_dataset.head()
```

```python
new_sample_dataset['Class'].value_counts()
```

```python
#checking the mean of new sample
new_sample_dataset.groupby('Class').mean()
```

```python
#checking the mean of old sample
input_df.groupby('Class').mean()
```

```python
#Separating the dependent and independent vairables
X=new_sample_dataset.drop(columns='Class',axis=1)
Y=new_sample_dataset['Class']
```

```python
#Checking the shape of newly created variables
print(X.shape)
print(Y.shape)
```

```python
#Splitting the dataset in to test and train dataset
X_train, X_test, Y_train, Y_test = train_test_split(X,Y,test_size=0.2,stratify=Y,random_state=2)
```

```python
#shape of all the test and train data
print(X_train.shape)
print(X_test.shape)
print(Y_train.shape)
print(Y_test.shape)
```

```python
#Defining the logistic regression model
model = LogisticRegression()
```

```python
# Training the logistic regression model with training data
model.fit(X_train,Y_train)
```

```python
#Model Evaluation
#Based on accuracy score
#Accuracy on training data
X_train_prediction = model.predict(X_train)
training_data_accuracy = accuracy_score(X_train_prediction,Y_train)
```

```python
print('Accuracy on training data :', training_data_accuracy)
```

```python
#Model Evaluation
#Based on accuracy score
#Accuracy on training data
X_test_prediction = model.predict(X_test)
testing_data_accuracy = accuracy_score(X_test_prediction,Y_test)
```

```python
print('Accuracy on testing data :', testing_data_accuracy)
```
