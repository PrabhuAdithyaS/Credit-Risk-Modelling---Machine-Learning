### Categorical Features Treatment

```python
emp_drp.head()
```

```python
#resetting the index
emp_drp.reset_index(drop=True,inplace=True)
```

```python
#checking the index after resetting the index
emp_drp.head()
```

```python
#creating a copy of the data
emp_dummy = emp_drp.copy()
emp_dummy.head()
```

```python
#Creating one hot one hot encoding for the categorical variables
person_home_ownership = pd.get_dummies(emp_dummy['person_home_ownership'],drop_first=True).astype(int)
loan_intent = pd.get_dummies(emp_dummy['loan_intent'],drop_first=True).astype(int)
emp_dummy['cb_person_default_on_file_binary'] = np.where(emp_dummy['cb_person_default_on_file']=='Y',1,0)
emp_dummy.head()
```

```python
person_home_ownership.head()
```

```python
#dropping the column for which we have created 1 and 0 instead of N and Y
emp_dummy.drop(columns='cb_person_default_on_file',inplace=True)
emp_dummy
```

```python
#coumns to standardize
data_to_scale = emp_dummy.drop(columns=['person_home_ownership','loan_intent','loan_status','cb_person_default_on_file_binary'])
data_to_scale.head()
```

```python
#standardizing the values
scaler = StandardScaler()
scaled_data = scaler.fit_transform(data_to_scale)
scaled_data
```

```python
#to find the shape of the array after scaling
scaled_data.shape
```

```python
data_to_scale.columns
```

```python
scaled_df = pd.DataFrame(scaled_data,columns=['person_age', 'person_income', 'person_emp_length', 'loan_amnt',
       'loan_int_rate', 'loan_percent_income', 'cb_person_cred_hist_length'])
scaled_df.head()
```

```python
#check the shape of the dataset after the standardization
scaled_df.shape
```

```python
#checking the mean (should be close to 0) and the std (should be close to 1) to confirm the values are scaled
print(round(scaled_df['person_age'].std(),0))
print(round(scaled_df['person_age'].mean(),0))
```

```python
#concatenating the requied columns (bringing in both scaled numerical columns and categorical columns
scaled_df_comb = pd.concat([scaled_df,person_home_ownership,loan_intent],axis=1)
scaled_df_comb.head()
```

```python
scaled_df_comb.shape
```

```python
#getting the binary column created 
scaled_df_comb['cb_person_default_on_file'] = emp_dummy['cb_person_default_on_file_binary']
```

```python
scaled_df_comb.shape
```

```python
scaled_df_comb.head()
```

```python
#getting the dependent variable
scaled_df_comb['loan_status'] = emp_dummy['loan_status']
scaled_df_comb.head()
```

```python
#shape of the output table created
scaled_df_comb.shape
```

```python
#checking whether the data as imbalanced or balanced (It has to be either 50:50 or 60:40)
scaled_df_comb['loan_status'].value_counts()
```

```python
#handling imbalanced data
smote = SMOTE()
```

```python
#getting the dependent and the independent variables separately
x = scaled_df_comb.drop(columns='loan_status')
y = scaled_df_comb['loan_status']
```

```python
#checking the target variable
x.head()
```

```python
#checking the feature variable
y.head()
```

```python
#careting the over sampled output
x_smote, y_smote = smote.fit_resample(x,y)
```

```python
#comparing the shapes before and after oversamopling - target variable
print(x.shape)
print(x_smote.shape)
```

```python
#comparing the shapes before and after oversamopling - feature variable
print(y.shape)
print(y_smote.shape)
```

```python
#checking the distribution of target variable class
y_smote.value_counts()
```

```python
#Model Training
y_smote.value_counts()
```

```python
balanced_target_df = pd.DataFrame(y_smote)
balanced_target_df.value_counts()
```

```python
x_train,x_test,y_train,y_test = train_test_split(x_smote,balanced_target_df,test_size=0.2,random_state=42)
```

```python
print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
logit = LogisticRegression()
```

```python
logit.fit(x_train,y_train)
```

```python
logit.score(x_train,y_train)
```

```python
logit_prediction = logit.predict(x_test)
logit_prediction
```

```python
print(classification_report(y_test,logit_prediction))
```

```python
print(logit.coef_[0])
```

```python
feature_importance = pd.DataFrame({'Features':x_train.columns,'Importance':logit.coef_[0]})
feature_importance.sort_values(by='Importance',ascending=False)
```

```python
#Random Forest
```

```python
rf = RandomForestClassifier()
```

```python
rf_fit = rf.fit(x_train,y_train)
```

```python
rf.score(x_train,y_train)
```

```python
rf_prediction = rf.predict(x_test)
rf_prediction
```

```python
print(classification_report(y_test,rf_prediction))
```

```python
rf.feature_importances_
```

```python
feature_importance_rf = pd.DataFrame({'Features_rf':x_train.columns,'Importance_rf':rf.feature_importances_})
feature_importance_rf.sort_values(by='Importance_rf',ascending=False)
```

## XG Boost

```python
xgb_model = XGBClassifier(tree_method='exact')
```

```python
xgb_model.fit(x_train,y_train.values.ravel())
```

```python
xgb_model.score(x_train,y_train.values.ravel())
```

```python
xgb_prediction = xgb_model.predict(x_test)
xgb_prediction
```

```python
print(classification_report(y_test,xgb_prediction))
```

```python
feature_importance_xgb = pd.DataFrame({'Features_xgb':x_train.columns,'Importance_xgb':xgb_model.feature_importances_})
feature_importance_xgb.sort_values(by='Importance_xgb',ascending=False)
```

```python
feature_imp_final = feature_importance.merge(feature_importance_rf,how='inner',left_on='Features',right_on='Features_rf') \
                                       .merge(feature_importance_xgb,how='inner',left_on='Features',right_on='Features_xgb') 
feature_imp_final.drop(['Features_rf','Features_xgb'],axis=1,inplace=True)
```
